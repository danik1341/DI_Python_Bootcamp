```python
x = np.linspace(0, 2 * np.pi, 100) 
y1 = np.sin(x)
y2 = np.cos(x)

figure, axes = plt.subplots(2)

axes[0].plot(x, y1, color='tab:blue', label='y = sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Line Plot of sin(x)')
axes[0].legend()

axes[1].scatter(x, y2, color='red', marker='o', label='y = cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Scatter Plot of cos(x)')
axes[1].legend()

plt.tight_layout()


## Kinda did 7 and 8 together .... and 9 it seems so
```

## 1

```python
df = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Temperature': [22, 24, 23, 24, 25, 26, 25],
    'Rainfall': [1, 0, 2, 1, 0, 0, 1]
})
df.set_index('Day', inplace=True)

df.plot(kind='line')

plt.xlabel('Day')
plt.ylabel('Value')
plt.title('Daily Temperature and Rainfall')
```

## 2

```python
std_dev = 1
df.plot(kind='line', yerr=std_dev)
plt.xlabel('Day')
plt.ylabel('Value')
plt.title('Daily Temperature and Rainfall with Error Bars')
```

## 3

```python
df = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Temperature': [22, 24, 23, 24, 25, 26, 25],
    'Rainfall': [1, 0, 2, 1, 0, 0, 1]
})

x = np.arange(len(df))

fig, ax1 = plt.subplots()

ax1.plot(x, df['Temperature'], color='tab:blue', label='Temperature')
ax1.set_xlabel('Day')
ax1.set_ylabel('Temperature', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()

ax2.plot(x, df['Rainfall'], color='tab:red', label='Rainfall')
ax2.set_ylabel('Rainfall', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

plt.title('Temperature and Rainfall on Twin Axes')
```

```python
df = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Temperature': [22, 24, 23, 24, 25, 26, 25],
    'Rainfall': [1, 0, 2, 1, 0, 0, 1]
})

df.set_index('Day', inplace=True)

ax = df.plot(kind='bar', color=['tab:blue', 'tab:red'], edgecolor = 'black')

ax.set_title('Temperature and Rainfall')
ax.set_xlabel('Day')
ax.set_ylabel('Value')
```

## 5

```python
titanic = sns.load_dataset('titanic')
plt.ion()

columns_to_plot = ['age', 'fare']

for column in columns_to_plot:
    fig, ax = plt.subplots()
    sns.histplot(data=titanic.dropna(subset=[column]), x=column, kde=True, ax=ax)
    ax.set_title(f'Distribution of {column}')
    plt.draw()
    plt.pause(1)
    plt.clf()

plt.ioff()
```
