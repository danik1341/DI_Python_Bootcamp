## 4

```python
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

figure, axes = plt.subplots()
plt.grid(True)

axes.set_xlabel('Price')
axes.set_ylabel('Demand')
axes.set_title('Product Demand vs Price')

axes.plot(x, y)

figure.savefig('Demand_Price.png') ## ex 6
```

## 5

## 7

```python
x = np.linspace(0, 2 * np.pi, 100) 
y1 = np.sin(x)
y2 = np.cos(x)

figure, axes = plt.subplots(2)

axes[0].plot(x, y1, color='tab:blue', label='y = sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('Line Plot of sin(x)')
axes[0].legend()

axes[1].scatter(x, y2, color='red', marker='o', label='y = cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('Scatter Plot of cos(x)')
axes[1].legend()

plt.tight_layout()


## Kinda did 7 and 8 together .... and 9 it seems so
```
