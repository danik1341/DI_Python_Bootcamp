```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 100) 
y = x ** 2

figure, axes = plt.subplots()

# axes.set_xlim(0, 20) it doenst look nice now :(
axes.set_xlabel('x')

# axes.set_xlim(0, 400) it doenst look nice now :(
axes.set_ylabel('y')

axes.set_title('Plot of x vs y (y=x^2)')

axes.plot(x, y)

figure.savefig('xvs.png')
```

## 2

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep='\t')
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
chipo['item_price']
```

```python
x = chipo['item_price']
y = chipo['quantity']

figure, axes = plt.subplots()
plt.grid(True)

axes.set_xlabel('Item Price')
axes.set_ylabel('Quantity')
axes.set_title('Relationship between Item Price and Quantity')

axes.scatter(x, y)

figure.savefig('price_quty.png')
```

## 3

```python
item_quantity = chipo.groupby('item_name')['quantity'].sum()

top_items = item_quantity.sort_values(ascending=False).head(5)

figure, axes = plt.subplots()
axes.bar(top_items.index, top_items.values, edgecolor = 'black')
axes.set_xlabel('Item Name')
axes.set_ylabel('Total Quantity')
axes.set_title('Top 5 Items by Total Quantity')

axes.set_xticklabels(top_items.index, rotation=45)

figure.savefig('top5.png')
```
